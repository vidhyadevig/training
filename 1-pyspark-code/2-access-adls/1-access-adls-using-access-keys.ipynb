#### Access Azure Data Lake using access keys
- Set the spark config for access key
- List files from the container
- Read data from the csv file

```python
# Step 1: Define your storage account and container
storage_account_name = "sanewtest567"
container_name = "raw"

# Step 2: Define the access key
access_key = ""

# Step 3: Configure Spark to use the access key
spark.conf.set(
    f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net",
    access_key
)

# Step 4: Define the ADLS path
adls_path = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/"

# List the files in the ADLS path
display(dbutils.fs.ls(adls_path))

# Step 5: Example - Read a CSV file from ADLS
df = spark.read.option("header", "true").csv(adls_path + "csv/employees.csv")

# Step 6: Display data
display(df)
```
