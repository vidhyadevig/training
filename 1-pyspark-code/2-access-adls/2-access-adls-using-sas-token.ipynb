#### Access Azure Data Lake using SAS Token
- Set the spark config for SAS Token
- List files from the container
- Read data from the csv file

```python
# Step 1: Define your storage account and container
storage_account_name = "sanewtest567"
container_name = "raw"

# Step 2: Define the sas token
sas_token = ""

# Step 3: Configure Spark to use the SAS token
spark.conf.set("fs.azure.account.auth.type.{storage_account_name}.dfs.core.windows.net", "SAS")
spark.conf.set("fs.azure.sas.token.provider.type.{storage_account_name}.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.sas.FixedSASTokenProvider")
spark.conf.set("fs.azure.sas.fixed.token.{storage_account_name}.dfs.core.windows.net", sas_token)

# Step 4: Define the ADLS path
adls_path = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/"

display(dbutils.fs.ls(adls_path))
```

```python
# Step 5: Example - Read a CSV file from ADLS
df = spark.read.option("header", "true").csv(adls_path + "csv/employees.csv")

# Step 6: Display data
display(df)
```
